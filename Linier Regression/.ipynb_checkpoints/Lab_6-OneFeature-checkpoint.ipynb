<h3>Загрузка датасету</h3>

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

dataset = pd.read_csv('Salary_dataset.csv')
dataset.drop('Unnamed: 0', axis=1, inplace=True)
dataset
```

<h3>Розбиття ознак (один Х)</h3>

```python
# Одержання ознак (X) та цільової змінної (y)
X = np.array(dataset['Salary']).reshape(-1, 1)
y = np.array(dataset['YearsExperience']).reshape(-1, 1)
```

<h3>Розбиття на тренувальну та тестову вибірки</h3>

```python
from sklearn.model_selection import train_test_split

# Розбиття на тестову та тренувальну вибірки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

<h3>Створення моделі лінійної регресії та тренування</h3>

```python
from sklearn.linear_model import LinearRegression

# Створення та навчання моделі з однією функцією (години вивчення)
regression = LinearRegression()
regression.fit(X_train, y_train)
print(regression.score(X_test, y_test))

#Вивести а і b
print("Коефіцієнт 'a':", regression.coef_[0])
print("Коефіцієнт 'b':", regression.intercept_)
```

<h3>Прогнозування та візуалізація</h3>

```python
y_pred = regression.predict(X_test)
plt.scatter(X_test, y_test, color ='b')
plt.plot(X_test, y_pred, color ='k')
 
plt.show()
```

<h3>Середня абсолютна помилка (чим меньше - краще)</h3>

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import r2_score

mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
print("Mean Absolute Error:", mae)

# Оцінка моделі
mse = mean_squared_error(y_test, y_pred)

# Calculate R-squared
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error:', mse)
print('R-squared:', r2)
```

<h3>Підбір оптимальних параметрів, використовуючи Ridge</h3>

```python
from sklearn.linear_model import RidgeCV

# Create and fit the RidgeCV model
ridge = RidgeCV(alphas=[0.1, 1.0, 10.0])
ridge.fit(X_train, y_train)

# The best alpha found by RidgeCV
best_alpha_ridge = ridge.alpha_

# Print the best alpha
print("Best alpha for Ridge:", best_alpha_ridge)
print("Коефіцієнт 'a':", ridge.coef_[0])
print("Коефіцієнт 'b':", ridge.intercept_)
```

<h3>Підбір оптимальних параметрів, використовуючи Lasso</h3>

```python
from sklearn.linear_model import LassoCV

# Create and fit the LassoCV model
lasso = LassoCV(alphas=[0.1, 1.0, 10.0])
lasso.fit(X_train, y_train.ravel())

# The best alpha found by LassoCV
best_alpha_lasso = lasso.alpha_

# Print the best alpha
print("Best alpha for Lasso:", best_alpha_lasso)
print("Коефіцієнт 'a':", lasso.coef_[0])
print("Коефіцієнт 'b':", lasso.intercept_)
```

<h3>Візуалізація оптимальних моделей</h3>

```python
# Visualize the initial model
plt.scatter(X_test, y_test, color='b', label='Actual Data')
plt.plot(X_test, y_pred, color='k', label='Initial Linear Regression')
plt.xlabel('Salary')
plt.ylabel('Years of Experience')
plt.legend()
plt.show()

# Visualize Ridge model
ridge_pred = ridge.predict(X_test)
plt.scatter(X_test, y_test, color='b', label='Actual Data')
plt.plot(X_test, ridge_pred, color='r', label=f'Ridge Regression (alpha={best_alpha_ridge})')
plt.xlabel('Salary')
plt.ylabel('Years of Experience')
plt.legend()
plt.show()
# Visualize Lasso model
lasso_pred = lasso.predict(X_test)
plt.scatter(X_test, y_test, color='b', label='Actual Data')
plt.plot(X_test, lasso_pred, color='g', label=f'Lasso Regression (alpha={best_alpha_lasso})')
plt.xlabel('Salary')
plt.ylabel('Years of Experience')
plt.legend()
plt.show()
```
