<h3>Загрузка датасету та бібліотек</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
 
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
```

```python
bbox = {'boxstyle' : 'round',
        'pad' : 0.5,
        'facecolor' : 'white',
        'edgecolor' : 'black',
        'linewidth' : 1,
        'alpha' : 1}

plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('wine-clustering.csv')
print(df.shape)
df.head()
```

```python
import seaborn as sns
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Set the size of the heatmap
plt.figure(figsize=(12, 10))

# Create a heatmap with seaborn
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Add a title
plt.title("Correlation Matrix", fontsize=16)

# Show the plot
plt.show()
```

<h3>Стандартизація даних, щоб вони були в одному діапазоні</h3>

```python
columns = df.columns
df[columns] = scaler.fit_transform(df[columns])
df.head()
```

```python
plt.scatter(df['Alcohol'], df['Flavanoids'])
plt.show()
```

```python
# List for silhouette scores
silhouette_scores = []

# Training the clustering model with different values of k
for k in range(2, 11):  # AgglomerativeClustering should have at least 2 clusters
    hierarchical_cluster = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='ward')
    labels = hierarchical_cluster.fit_predict(df)
    silhouette_avg = silhouette_score(df, labels)
    silhouette_scores.append(silhouette_avg)

# Plot the silhouette scores
plt.figure(figsize=(12, 8))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Selecting the Optimal Number of Clusters', fontsize=14)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 11))
plt.grid(True)
plt.show()
```

```python
linkage_data = linkage(df, method='ward', metric='euclidean')
dendrogram(linkage_data, truncate_mode='lastp', p=12)
plt.figure(figsize=(12, 8))
plt.show()
```

```python
hierarchical_cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
labels = hierarchical_cluster.fit_predict(df)

# You should select two features for visualization; for example, let's use 'Alcohol' and 'Malic_Acid'
feature1 = 'Alcohol'
feature2 = 'Flavanoids'

# Visualize the clusters using a scatter plot
plt.scatter(df[feature1], df[feature2], c=labels, cmap='viridis')
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.title('Agglomerative Clustering')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(df, labels)
print(f"Silhouette Score: {silhouette_avg}")
```

```python
from sklearn.metrics import davies_bouldin_score
db_index = davies_bouldin_score(df, labels)
print(f"Davies-Bouldin Index: {db_index}")
```

```python
from sklearn.metrics import calinski_harabasz_score
ch_index = calinski_harabasz_score(df, labels)
print(f"Calinski-Harabasz Index: {ch_index}")
```
