<h3>Загрузка датасету з файлу CSV</h3>

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns


# Зчитування CSV файлу та вивід даних
#data = pd.read_csv('spotify-2023-main.csv', delimiter=',', encoding='latin1')
data = pd.read_csv('spotify_normalized_data.csv', delimiter=',', encoding='latin1')

data.drop('Unnamed: 0', axis=1, inplace=True)
data
```

<h3>Групування даних</h3>

```python
grouped_data = data[['key', 'mode','streams']]
grouped_data = data.groupby(['key', 'mode'], as_index=False)['streams'].mean(numeric_only=True)
grouped_data
```

<h3>Сортування даних</h3>

```python
sorted_grouped_data = grouped_data.sort_values(by='streams', ascending=False)
sorted_grouped_data
```

<h3>Таблиця спряженості для key пісень в Spotify та кількістью streams</h3>

```python
pd.crosstab(data['key'], data['streams'])
```

<h3>Функція для виводу гістограми, Q-Q plot, Boxplot</h3>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

def diagnostic_plots(df, variable, title):
    fig, ax = plt.subplots(figsize=(10, 7))
    
    # Гістограма
    plt.subplot(2, 2, 1)
    df[variable].hist()
    
    # Q-Q plot
    plt.subplot(2, 2, 2)
    stats.probplot(df[variable], dist='norm', plot=plt)
    
    # Boxplot
    plt.subplot(2, 2, 3)
    sns.boxplot(x = df[variable])
    
    fig.suptitle(title)
    plt.show()
```

<h3>Вивід нормального розподілу року випуску пісні</h3>

```python
diagnostic_plots(data, 'streams', 'in_apple_charts - original')
```

<h3>К-тест</h3>

```python
from scipy.stats import kstest, norm
import numpy as np

ks_statistic, ks_p_value = kstest(data['streams'], 'norm')

alpha = 0.05

# Перевірка 
if ks_p_value < alpha:
    print("Не нормальний розподіл")
else:
    print("Нормальний розподіл")

# Вивід
print("K-S Statistic:", ks_statistic)
print("P-value:", ks_p_value)
```

<h3>Вивід після логарифмування</h3>

```python
alternative_data = data

alternative_data['in_apple_charts_log'] = np.log(alternative_data['streams'] + 11)
diagnostic_plots(alternative_data, 'in_apple_charts_log', 'in_apple_charts after Log')

# Знову за К-тестом перевірка
ks_statistic, ks_p_value = kstest(alternative_data['in_apple_charts_log'], 'norm')

alpha = 0.05

# Перевірка 
if ks_p_value < alpha:
    print("Не нормальний розподіл")
else:
    print("Нормальний розподіл")

# Вивід
print("K-S Statistic:", ks_statistic)
print("P-value:", ks_p_value)
```

<h3>Вивід після степеневого перетворення</h3>

```python
alternative_data = data

# Степеневе перетворення ознаки 
alternative_data['power_transform_in_apple_charts'] = alternative_data['streams'] ** 5
diagnostic_plots(alternative_data, 'power_transform_in_apple_charts', 'in_apple_charts after Power Transformed')

# Знову за К-тестом перевірка
ks_statistic, ks_p_value = kstest(alternative_data['power_transform_in_apple_charts'], 'norm')

alpha = 0.05

# Перевірка 
if ks_p_value < alpha:
    print("Не нормальний розподіл")
else:
    print("Нормальний розподіл")

# Вивід
print("K-S Statistic:", ks_statistic)
print("P-value:", ks_p_value)
```

<h3>Перетворення Yeo-Johnson</h3>

```python
from scipy.stats import yeojohnson

alternative_data = data

# Застосування Yeo-Johnson трансформації
alternative_data['in_apple_charts_yeojohnson'], _ = yeojohnson(alternative_data['in_apple_charts'])

diagnostic_plots(alternative_data, 'in_apple_charts_yeojohnson', 'in_apple_charts after BoxCox')

# Знову за К-тестом перевірка
ks_statistic, ks_p_value = kstest(alternative_data['in_apple_charts_yeojohnson'], 'norm')

alpha = 0.05

# Перевірка 
if ks_p_value < alpha:
    print("Не нормальний розподіл")
else:
    print("Нормальний розподіл")

# Вивід
print("K-S Statistic:", ks_statistic)
print("P-value:", ks_p_value)
```

<h3>Перетворення Box-Cox</h3>

```python
from scipy.stats import boxcox
import numpy as np

alternative_data = data

# Perform the Box-Cox transformation
const = 10
released_year_data = alternative_data['in_apple_charts'] + const
transformed_data, best_lambda = boxcox(released_year_data)

alternative_data['boxcox_transformed_in_apple_charts'] = np.where(alternative_data['in_apple_charts'].notnull(), transformed_data, np.nan)


ks_statistic, ks_p_value = kstest(alternative_data['boxcox_transformed_in_apple_charts'], 'norm')

alpha = 0.05


if ks_p_value < alpha:
    print("Не нормальний розподіл")
else:
    print("Нормальний розподіл")
    

print("K-S Statistic:", ks_statistic)
print("P-value:", ks_p_value)
```

<h3>Feature Engineering</h3>

```python
alternative_data = data

# Створення нової ознаки 'is_popular'
alternative_data['is_popular'] = (alternative_data['in_spotify_playlists'] > 0) | (alternative_data['in_spotify_charts'] > 0)

# Вивід перших рядків оновленого набору даних
grouped_data = alternative_data.groupby(['in_spotify_playlists', 'in_spotify_charts'], as_index=False)['is_popular'].mean()

# Display the grouped data
data = alternative_data 

data
```

<h3>Перетворення одних типів значень в інші</h3>

```python
from sklearn.preprocessing import LabelEncoder

# Перетворення категоріальної ознаки 'artist(s)_name' у числову за допомогою Label Encoding
label_encoder = LabelEncoder()
alternative_data['artist(s)_name'] = label_encoder.fit_transform(alternative_data['artist(s)_name'])

# Перетворення числової ознаки 'released_month' у категоріальну
alternative_data['released_month'] = alternative_data['released_month'].astype('category')

print(alternative_data.info())

# Вивід перших рядків оновленого набору даних
alternative_data[['artist(s)_name', 'released_month']].head()
```
